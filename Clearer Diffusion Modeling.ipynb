# DIFFUSION MODELING

##### using the diffusion equation to measure how many cosmic rays will make its way to earth from SN at different distances

n(E, r ,t )=  Q(E) / ((np.pi**(3/2)) * (rdiff**3) * exp((-r**2)/(rdiff**2)) 

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
def diffusion(E,r,t, **kwargs): #defining the diffusion equation
    Q  = 10e50 * ((E**-2.2)* np.exp((-E)/1e6))
    D  = (7*10e-2) * (E/(10))**(1/3)
    rd = (4 * D * t) **.5
    n  = (Q/ (((np.pi**(3/2)) * (rd**3)))) * (np.exp(-(r**2) / rd**2))
    return n
```

```python
#trying a for loop so that i dont have to type everything out everytime for each graph...
#constant time and distance

x_coords = np.linspace(10,50e5) #x coords E are defined from 1 GeV to 20e5 with 20e5 intervals inbetween
plt.figure(figsize=(10, 10)) # figure size
for i in range(1,10): #i in range time
    y_coords = diffusion(x_coords,100, 10**i) #50 pc, i is the incriments of time
    plt.plot((np.log10(x_coords)),(np.log10(y_coords)), alpha = .5, linewidth=3.0, label = f'10^{i} years') #plotting the graphs
    name = 'plot' + str(i) #naming the string 

    #print(np.log10(10**i))
    #print(np.log10(y_coords))
plt.title("Diffusion Distribution 1000 GeV")
plt.xlabel("Energy(GeV)")
plt.ylabel("number of particles(n)")
plt.vlines(5, 0 ,40, color = 'black', label = 'energy cuttoff', linestyle = '--')
plt.ylim(0,40)
plt.legend()
plt.show()

```
